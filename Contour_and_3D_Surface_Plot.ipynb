```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./4K_House_Rent_Dataset.csv')
X = data['Size']
Y = data['Rent']

m = len(X)
```

```python
# Define the loss function
def loss(a, b):
    return 1 / m * np.sum((a * X + b - Y) ** 2)
```

```python
# Generate values for parameters a and b
a_values = np.linspace(-400, 400, 100)
b_values = np.linspace(-300000, 300000, 100)

A, B = np.meshgrid(a_values, b_values)
```

```python
# Calculate loss values for each combination of a and b
loss_vals = np.zeros_like(A)
for i in range(len(a_values)):
    for j in range(len(b_values)):
        loss_vals[i, j] = loss(a_values[i], b_values[j])
```

```python
# Contour plot
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
contour_plot = plt.contour(A, B, loss_vals, levels=np.linspace(0, 2e11, 20), cmap='viridis')
plt.colorbar(contour_plot, label='Loss')
plt.xlabel('a')
plt.ylabel('b')
plt.title('Contour Plot of Loss')
```

```python
# 3D Surface plot
plt.figure(figsize=(16, 6))

ax = plt.subplot(1, 2, 2, projection="3d")
surface_plot = ax.plot_surface(A, B, loss_vals, cmap="viridis", rstride=3, cstride=3, edgecolor='none', alpha=0.8)
plt.colorbar(surface_plot, label='Loss')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('Loss')
plt.title('3D Surface Plot of Loss')
```
